```python
import pandas as pd

# Ensure 'rets' is defined before using it
if 'rets' not in locals():
    rets = np.log(data_all / data_all.shift(1))
    rets = rets.dropna()

avg_rets = rets.mean() * 252    # annualized average returns

cov_matrix = rets.cov() * 252   # annualized covariance matrix

avg_vol = rets.std() * np.sqrt(252)  # annualized volatility   
sharpe_ratio =  (avg_rets - 0.01) / avg_vol # assuming risk-free rate of 1%

# Create a dataframe with the results
result_df = pd.DataFrame({
    'Average Returns': avg_rets,
    'Average Volatility': avg_vol,
    'Sharpe Ratio': sharpe_ratio
})

# Output the results into an Excel file
%pip install openpyxl
result_df.to_excel('results.xlsx')

# Output the covariance matrix into a separate Excel file
cov_matrix.to_excel('cov_matrix.xlsx')

result = (avg_rets, avg_vol, cov_matrix, sharpe_ratio, result_df)
```
