```python
import yfinance as yf
start_date = '1990-1-1'
end_date ='2024-3-31'


spx_all= yf.download('^GSPC', start=start_date, end=end_date)
vix_all = yf.download('^VIX', start=start_date, end=end_date)

spxvix3 = pd.DataFrame({'SPX': spx_all['Close'], 'VIX': vix_all['Close']}, index=spx_all.index)
spxvix3.plot(subplots=True, figsize=(10, 6),color=['blue','red'])
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:blue'
ax1.set_xlabel('Date')
ax1.set_ylabel('SPX', color=color)
ax1.plot(spxvix3.index, spxvix3['SPX'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('VIX', color=color)  # we already handled the x-label with ax1
ax2.plot(spxvix3.index, spxvix3['VIX'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
